# **NAVVTC TASKS**

```python
# Addition of integers
def integer_addition(a, b):
    return a + b

a = int(input("Enter an integer: ").strip())
b = int(input("Enter an integer: ").strip())

result = integer_addition(a, b)

print(f"Sum of integers {a} and {b}: {result}")
```

```python
# Addition of floats
def float_addition(a, b):
    return a + b
a = float(input("Enter a float: ").strip())
b = float(input("Enter a float: ").strip())

result = float_addition(a, b)
print("Sum of floats", a, " and ", b, " : ", result)
```

```python
# Calculate Square root of a real number
def calculate_sqroot(n):
    return n ** 0.5

n = float(input("Enter a number: ").strip())

result = calculate_sqroot(n)

print("Square root of ", n, " : ", result)
```

```python
# Square root of a complex number
import cmath

complex_number = eval(input("Enter a complex number (in the form a+bj): ").strip())

result = cmath.sqrt(complex_number)

print("Square root of ", complex_number, " : ", result)
```

```python
# Convert Fahrenheit to Celsius 
def convert_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5.0/9.0

fahrenheit = float(input("Enter temperature in Fahrenheit: ").strip())

celsius = convert_to_celsius(fahrenheit)

print(f"{fahrenheit} degrees Fahrenheit is equal to {celsius} degrees Celsius.")
```

```python
# Area of Triangle
def calculate_triangle_area(base, height):
    return 0.5 * base * height

base = float(input("Enter the base of the triangle: ").strip())
height = float(input("Enter the height of the triangle: ").strip())

area = calculate_triangle_area(base, height)

print(f"The area of the triangle with base {base} and height {height} is: {area}")
```

```python
# Solve a quadratic equation

import cmath

a = float(input("Enter coefficient a: ").strip())
b = float(input("Enter coefficient b: ").strip())
c = float(input("Enter coefficient c: ").strip())

discriminant = b**2 - 4*a*c

root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
root2 = (-b - cmath.sqrt(discriminant)) / (2*a)

print(f"The roots of the quadratic equation {a}x^2 + {b}x + {c} = 0 are: {root1} and {root2}")
```

```python
# Convert Kilometers to Miles
def convert_km_to_miles(km):
    return km * 0.621371

km = float(input("Enter distance in kilometers: ").strip())

miles = convert_km_to_miles(km)

print(f"{km} kilometers is equal to {miles} miles.")
```

# **FOR LOOP**

```python
# Sum of a list of numbers
def sum_of_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

numbers = [1, 2, 3, 4, 5]
result = sum_of_list(numbers)
print(f"The sum of the list {numbers} is: {result}")
```

# **WHILE LOOP**

```python
i = 0
while i < 4:
    i += 1
    print(i)
else:  # Executed because no break in for
    print("No Break\n")

i = 0
while i < 4:
    i += 1
    print(i)
    break
else:  # Not executed as there is a break
    print("No Break")
```

# **Break Statement**

```python
a = [1, 3, 5, 7, 9, 11]
val = 7

for i in a:
    if i == val:
        print(f"Found at {i}!")
        break
else:
    print(f"not found")
```

# **Continue Statement**

```python
for char in "muhammad":
    if char == "m":
        continue
    print(char, end=" ")
```
