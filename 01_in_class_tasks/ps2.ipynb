```python
### Quadratic formula Calculation
import cmath

def calculate_roots_using_quadratic_formula(a, b, c):
    if a == 0:
        print("Coefficient 'a' cannot be zero for a quadratic equation.")
        return None
    
    b_sqr_minus_four_a_c = (b**2) - (4)*(a)*(c)
    sqrt_b_sqr_expression = cmath.sqrt(b_sqr_minus_four_a_c)

    pos_root = ((-1)*(b) + (sqrt_b_sqr_expression)) / ((2) * (a))
    neg_root = ((-1)*(b) - (sqrt_b_sqr_expression)) / ((2) * (a))
    
    return (pos_root, neg_root)

a = int(input("Enter value for coefficient 'a': "))
b = int(input("Enter value for coefficient 'b': "))
c = int(input("Enter value for constant 'c': "))

result = calculate_roots_using_quadratic_formula(a,b,c)
if result != None:
    root1, root2 = result
    print(f"Roots of the equation are:\nx = {root1}\nx = {root2}")
```

```python
# Prime Number Check
def is_Prime(number):
    if number == 2 or number == 3:
        return True
    for i in range(2, number):
        if number % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
result = is_Prime(number)

if result == True:
    print("Number is Prime.\n")
else:
    print("Number is not Prime.\n")
```

```python
# Convert Temperature
def convert_temperature(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = convert_temperature(celsius)
print(f"Temperature in Fahrenheit: {fahrenheit}")
```

```python
def is_armstrong(number):
    psum = int(number)
    count = len(number)
    sum1 = 0
        
    for i in range(1,6):
        for j in range(count):
            sum1 = (int(number[j])**i) + sum1
        if sum1 == psum:
            return True
        else:
            sum1 = 0
    return False

number = input("Enter number: ").strip()
result = is_armstrong(number)
if result == True:
    print("Number is Armstrong.\n")
else:
    print("Number is not Armstrong.\n")
```

```python
#String Palindrom
def is_Palindrom(clause):
    c2 = clause[::-1]
    for i in range(len(clause)):
        if clause[i] != c2[i]:
            return False
    return True

clause = input("Enter Clause: ").strip().lower()
result = is_Palindrom(clause)
if result == True:
    print("String is Palindrom.\n")
else:
    print("String is not a Palindrom.\n")
```
